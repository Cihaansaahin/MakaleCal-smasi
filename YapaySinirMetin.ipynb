```python
from googleapiclient.discovery import build
from textblob import TextBlob  # TextBlob'u import ettik

# API anahtarƒ±nƒ±zƒ± buraya ekleyin
api_key = "AIzaSyC_ybONJ0lbnPg6Lmt5N6vdRoTc6C9jB4A"
youtube = build('youtube', 'v3', developerKey=api_key)

# Yorumlarƒ±nƒ± √ßekmek istediƒüiniz video ID'si
video_id = "0_fDiwCTJwc"

# Yorumlarƒ± almak
request = youtube.commentThreads().list(
    part="snippet",
    videoId=video_id,
    maxResults=100
)
response = request.execute()

# Yorumlarƒ±n duygu analizi
for item in response['items']:
    comment = item['snippet']['topLevelComment']['snippet']['textDisplay']
    
    # TextBlob ile duygu analizi
    blob = TextBlob(comment)
    sentiment = blob.sentiment.polarity  # Polarity: -1 (negatif) to 1 (pozitif)
    
    if sentiment > 0:
        sentiment_label = 'Pozitif'
    elif sentiment < 0:
        sentiment_label = 'Negatif'
    else:
        sentiment_label = 'N√∂tr'
    
    print(f"Yorum: {comment}\nDuygu: {sentiment_label}\n")
```
